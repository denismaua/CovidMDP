```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler

X = np.array(escola[['co_esc_x', 'co_esc_y']])
X = MinMaxScaler().fit_transform(X)
clustering = DBSCAN(eps = 0.005, min_samples=3, n_jobs=-1).fit(X)
```

```python
mask = escola['zona_esc'] > 300
```

```python
px.scatter(x=X[mask,0], y=X[mask,1], color=list(map(lambda x: str(x), clustering.labels_[mask])))
```
